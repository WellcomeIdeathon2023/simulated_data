```python
#Retreiving trial_ID from clinical.csv dataset downloaded from clinical.gov

#study_trials = clinical['NCT Number']
clinical1 = pd.read_csv('sch.csv')
```

```python

clinical = clinical1.drop_duplicates()
clinical= clinical.fillna(0)
```

```python

mental_health_condition = ['Anxiety','Schizophrenia','Autism']
clinical_trial= clinical[clinical['Condition'].isin(mental_health_condition)]
```

```python
ratings = []
# Generate a list of unique User_IDs
user_id = random.sample(range(1, 5000+ 1), 5000)
```

```python
for u_id in user_id:
    age = random.randint(18, 50)
    sex = random.choice(['Male', 'Female','others'])
    rating = random.randint(1, 5)
    location = random.choice(["Tokyo","New York City","London","Paris","Berlin","Moscow","Sydney","Rio de Janeiro","Cairo","Beijing","Istanbul"])
    condition = random.choice(['Anxiety','Schizophrenia','Autism'])
    dictionary= {
        'Condition': condition,
        'ID': u_id,
        'Age': age,
        'Sex': sex,
        'rating': rating,
        'location': location
    }

    ratings.append(dictionary)
```

```python
ratings
```

```python
#Converting list of dictionary into dataframe
df = pd.DataFrame(ratings)
df.shape
```

```python
#To download the csv
df.to_csv('user_ratings.csv', index=False)
clinical_trial.to_csv('clinical_trial.csv', index=False)
```
